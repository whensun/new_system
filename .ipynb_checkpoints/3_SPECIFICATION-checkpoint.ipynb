```python
# 1 - CREATE - 1_RAW - START

# TIME START - BEGIN
import time
TIME_START = time.time()
# TIME START - FINISH

# IMPORT - BEGIN
import logging, os, re
os.environ.update({'TF_CPP_MIN_LOG_LEVEL': '3', 'TF_ENABLE_ONEDNN_OPTS': '0'})
from docling.document_converter import DocumentConverter
# IMPORT - FINISH

# PATH - BEGIN
FOLDER_INPUT = '2_INPUT'
FOLDER_OUTPUT = '3_SPECIFICATION/1_RAW'
# PATH - FINISH

# FUNCTION - BEGIN
def FUNCTION_NATURAL_SORT_KEY(STRING_VALUE):
    return [int(EACH_TOKEN) if EACH_TOKEN.isdigit() else EACH_TOKEN.lower() for EACH_TOKEN in re.split(r'(\d+)', STRING_VALUE)]
# FUNCTION - FINISH

# CODE - BEGIN
for EACH_FILE_NAME in os.listdir(FOLDER_OUTPUT):
    if EACH_FILE_NAME.endswith('.txt'):
        os.remove(os.path.join(FOLDER_OUTPUT, EACH_FILE_NAME))

logging.getLogger('docling').setLevel(logging.WARNING)
logging.getLogger('tensorflow').setLevel(logging.ERROR)

for EACH_FILE_NAME in sorted(os.listdir(FOLDER_INPUT), key=FUNCTION_NATURAL_SORT_KEY):
    if EACH_FILE_NAME.lower().endswith('.pdf'):
        FILE_PDF = os.path.join(FOLDER_INPUT, EACH_FILE_NAME)
        FILE_OUTPUT = os.path.join(FOLDER_OUTPUT, os.path.splitext(EACH_FILE_NAME)[0] + '.txt')
        with open(FILE_OUTPUT, 'w', encoding='utf-8', newline='\n') as FILE_HANDLE:
            FILE_HANDLE.write(DocumentConverter().convert(FILE_PDF).document.export_to_markdown())
# CODE - FINISH

# TIME END - BEGIN
TIME_END = time.time()
DURATION_IN_SECOND = TIME_END - TIME_START
DAYS = int(DURATION_IN_SECOND // 86400)
DURATION_IN_SECOND %= 86400
HOURS = int(DURATION_IN_SECOND // 3600)
DURATION_IN_SECOND %= 3600
MINUTES = int(DURATION_IN_SECOND // 60)
DURATION_IN_SECOND %= 60
PARTS = []
if DAYS:
    PARTS.append(f'{DAYS} day(s)')
if HOURS:
    PARTS.append(f'{HOURS} hour(s)')
if MINUTES:
    PARTS.append(f'{MINUTES} minute(s)')
if DURATION_IN_SECOND or not PARTS:
    PARTS.append(f'{DURATION_IN_SECOND:.2f} second(s)')
print(' '.join(PARTS))
# TIME END - FINISH

# DONE - BEGIN
print('====\nDONE\n====')
# DONE - FINISH

# 1 - CREATE - 1_RAW - END
```

```python
# 2 - CREATE - SPECIFICATION_ALL.txt - START

# TIME START - BEGIN
import time
TIME_START = time.time()
# TIME START - FINISH

# IMPORT - BEGIN
import html, os, re
# IMPORT - FINISH

# PATH - BEGIN
FOLDER_INPUT = '3_SPECIFICATION/1_RAW'
FILE_OUTPUT = '3_SPECIFICATION/SPECIFICATION_ALL.txt'
# PATH - FINISH

# FUNCTION - BEGIN
def FUNCTION_NATURAL_SORT_KEY(STRING_VALUE):
    return [int(EACH_TOKEN) if EACH_TOKEN.isdigit() else EACH_TOKEN.lower() for EACH_TOKEN in re.split(r'(\d+)', STRING_VALUE)]
# FUNCTION - FINISH

# VARIABLE - BEGIN
LIST_REMOVE_CHAR = ['\u200b', '\u200c', '\u200d', '\ufeff']
LIST_SPACE_CHAR = ['\u00a0', '\u202f']
MAP_REPLACE = {'“': '"', '”': '"', '←': '<-', '→': '->', '■': '+', '<!-- image -->': '<image>', '<!-- formula-not-decoded -->': '<formula>'}
STRING_VALUE = ''
# VARIABLE - FINISH

# CODE - BEGIN
for EACH_FILE_NAME in sorted(os.listdir(FOLDER_INPUT), key=FUNCTION_NATURAL_SORT_KEY):
    if not EACH_FILE_NAME.endswith('.txt'):
        continue
    with open(os.path.join(FOLDER_INPUT, EACH_FILE_NAME), encoding='utf-8') as FILE_HANDLE:
        STRING_CONTENT = html.unescape(FILE_HANDLE.read().strip())
    for EACH_STRING_KEY, EACH_STRING_REPLACEMENT in MAP_REPLACE.items():
        STRING_CONTENT = STRING_CONTENT.replace(EACH_STRING_KEY, EACH_STRING_REPLACEMENT)
    STRING_CONTENT = re.sub(r'[\ue000-\uf8ff]', '', STRING_CONTENT)
    for EACH_CHAR_REMOVE in LIST_REMOVE_CHAR:
        STRING_CONTENT = STRING_CONTENT.replace(EACH_CHAR_REMOVE, '')
    for EACH_CHAR_SPACE in LIST_SPACE_CHAR:
        STRING_CONTENT = STRING_CONTENT.replace(EACH_CHAR_SPACE, ' ')
    STRING_CONTENT = ''.join(EACH_CHARACTER for EACH_CHARACTER in STRING_CONTENT if not 0xe000 <= ord(EACH_CHARACTER) < 0xf900)
    LIST_INDEX = [EACH_MATCH.start() for EACH_MATCH in re.finditer(r'## Chapter', STRING_CONTENT)]
    STRING_CONTENT = STRING_CONTENT[LIST_INDEX[0]:LIST_INDEX[-1]].strip()
    STRING_VALUE += STRING_CONTENT + '\n\n'

with open(FILE_OUTPUT, 'w', encoding='utf-8') as FILE_HANDLE:
    FILE_HANDLE.write(STRING_VALUE.strip())
# CODE - FINISH

# TIME END - BEGIN
TIME_END = time.time()
DURATION_IN_SECOND = TIME_END - TIME_START
DAYS = int(DURATION_IN_SECOND // 86400)
DURATION_IN_SECOND %= 86400
HOURS = int(DURATION_IN_SECOND // 3600)
DURATION_IN_SECOND %= 3600
MINUTES = int(DURATION_IN_SECOND // 60)
DURATION_IN_SECOND %= 60
PARTS = []
if DAYS:
    PARTS.append(f'{DAYS} day(s)')
if HOURS:
    PARTS.append(f'{HOURS} hour(s)')
if MINUTES:
    PARTS.append(f'{MINUTES} minute(s)')
if DURATION_IN_SECOND or not PARTS:
    PARTS.append(f'{DURATION_IN_SECOND:.2f} second(s)')
print(' '.join(PARTS))
# TIME END - FINISH

# DONE - BEGIN
print('====\nDONE\n====')
# DONE - FINISH

# 2 - CREATE - SPECIFICATION_ALL.txt - END
```
