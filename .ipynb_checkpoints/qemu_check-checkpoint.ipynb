```python
'''
Check the result.txt.
'''
# 1 / 1
# output

# IMPORT
import operator
import re

# FUNCTION
def canonical_reg(name: str) -> str:
    key = name.strip()
    return REG_ALIAS.get(key, key)

def _parse_int(value_str: str) -> int:
    s = value_str.strip().lower()
    if s.startswith(('+0x', '-0x', '0x')):
        return int(value_str, 16)
    return int(value_str, 10)

def parse_iterations(content):
    pattern = re.compile(r'START ITER (\d+)(.*?)END ITER \1', re.DOTALL)
    iterations = pattern.findall(content)
    return iterations

def has_error(iter_content: str) -> bool:
    return re.search(r'\bpriv\b', iter_content) is None

def parse_expected_results(iter_content):
    value_re = r'[+-]?(?:0x[0-9a-fA-F]+|\d+)'
    reg_re = r'[A-Za-z_]\w+'
    rhs_re = rf'(?:{value_re}|{reg_re})'
    result_line_re = re.compile(r'#\s*Result:\s*(.*)')
    cond_re = re.compile(rf'\s*({reg_re})\s*(==|=|!=|<=|>=|<|>)\s*({rhs_re})\s*')
    expected = []
    for line in iter_content.splitlines():
        m = result_line_re.search(line)
        if not m:
            continue
        tail = m.group(1)
        for part in tail.split('&'):
            part = part.strip()
            if not part:
                continue
            m2 = cond_re.fullmatch(part)
            if not m2:
                continue
            reg, op, rhs_token = m2.groups()
            reg = canonical_reg(reg)
            if op == '=':
                op = '=='
            expected.append((reg, op, rhs_token))
    return expected

def parse_actual_registers(iter_content):
    value_re = r'[+-]?(?:0x[0-9a-fA-F]+|\d+)'
    reg_pattern = re.compile(rf'(\w+)\s+0x[0-9a-fA-F]+\s+({value_re})')
    actual = {}
    for reg, value in reg_pattern.findall(iter_content):
        try:
            v = _parse_int(value)
        except ValueError:
            continue
        canon = canonical_reg(reg)
        actual[canon] = v
        actual[reg] = v
    return actual

# BONUS
RESULT_ONLY_ERROR_RE = re.compile(r'^\s*#\s*Result\s*:\s*error\s*$', re.IGNORECASE | re.MULTILINE)

def has_result_only_error(iter_content: str) -> bool:
    return RESULT_ONLY_ERROR_RE.search(iter_content) is not None

def check_iterations(iterations):
    string = ''
    for iter_num, iter_content in iterations:
        iter_num = int(iter_num)
        if has_result_only_error(iter_content):
            if has_error(iter_content):
                continue
            else:
                var = f'START ITER {iter_num} is false'
                print(var)
                string += var + '\n'
                continue
        if has_error(iter_content):
            var = f'START ITER {iter_num} is error'
            print(var)
            string += var + '\n'
            continue
        expected_list = parse_expected_results(iter_content)
        actual = parse_actual_registers(iter_content)
        mismatch = False
        for reg, op, rhs_token in expected_list:
            reg_key = canonical_reg(reg)
            act_val = actual.get(reg_key)
            if act_val is None:
                mismatch = True
                break
            try:
                exp_val = _parse_int(rhs_token)
            except ValueError:
                rhs_key = canonical_reg(rhs_token)
                rhs_val = actual.get(rhs_key)
                if rhs_val is None:
                    mismatch = True
                    break
                exp_val = rhs_val
            if op in ('==', '!='):
                a = u64(act_val)
                e = u64(exp_val)
                if not OPS[op](a, e):
                    mismatch = True
                    break
            else:
                if not OPS[op](act_val, exp_val):
                    mismatch = True
                    break
        if mismatch:
            var = f'START ITER {iter_num} is false'
            print(var)
            string += var + '\n'
    string_finalized = string.strip()
    with open(output, 'w', encoding='utf-8') as file:
        file.write(string_finalized)

def u64(x: int) -> int:
    return x & MASK64

# PATH
filename = 'qemu_result.txt'
output = 'qemu_check.txt'

# VARIABLE
OPS = {
    '=':  operator.eq,
    '==': operator.eq,
    '!=': operator.ne,
    '<':  operator.lt,
    '<=': operator.le,
    '>':  operator.gt,
    '>=': operator.ge,
}

REG_ALIAS = {
    'x0': 'x0', 'zero': 'x0',
    'x1': 'x1', 'ra': 'x1',
    'x2': 'x2', 'sp': 'x2',
    'x3': 'x3', 'gp': 'x3',
    'x4': 'x4', 'tp': 'x4',
    'x5': 'x5', 't0': 'x5',
    'x6': 'x6', 't1': 'x6',
    'x7': 'x7', 't2': 'x7',
    'x8': 'x8', 's0': 'x8', 'fp': 'x8',
    'x9': 'x9', 's1': 'x9',
    'x10': 'x10', 'a0': 'x10',
    'x11': 'x11', 'a1': 'x11',
    'x12': 'x12', 'a2': 'x12',
    'x13': 'x13', 'a3': 'x13',
    'x14': 'x14', 'a4': 'x14',
    'x15': 'x15', 'a5': 'x15',
    'x16': 'x16', 'a6': 'x16',
    'x17': 'x17', 'a7': 'x17',
    'x18': 'x18', 's2': 'x18',
    'x19': 'x19', 's3': 'x19',
    'x20': 'x20', 's4': 'x20',
    'x21': 'x21', 's5': 'x21',
    'x22': 'x22', 's6': 'x22',
    'x23': 'x23', 's7': 'x23',
    'x24': 'x24', 's8': 'x24',
    'x25': 'x25', 's9': 'x25',
    'x26': 'x26', 's10': 'x26',
    'x27': 'x27', 's11': 'x27',
    'x28': 'x28', 't3': 'x28',
    'x29': 'x29', 't4': 'x29',
    'x30': 'x30', 't5': 'x30',
    'x31': 'x31', 't6': 'x31',
}

MASK64 = (1 << 64) - 1

# CODE
with open(filename, 'r', encoding='utf-8') as file:
    content = file.read()

iterations = parse_iterations(content)
check_iterations(iterations)

# DONE
print('====\nDONE\n====')

# 1 / 1
# output
```
