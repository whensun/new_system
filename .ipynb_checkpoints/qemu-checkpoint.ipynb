```python
'''
Generate the result.txt.
'''
# 1 / 1
# result.txt

# IMPORT
from pwn import *
import re
import subprocess
import sys

# FUNCTION
def parse_registers(file_path):
    registers = {}
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            parts = line.split()
            if len(parts) >= SKIPPING_N_LINES:
                registers[parts[0]] = parts[1]
    return registers

def clean_register_output(output):
    lines = output.splitlines()
    trimmed_output = '\n'.join(lines[SKIPPING_N_LINES:-1])
    return ANSI_ESCAPE.sub('', trimmed_output)

def handle_qemu_error(cmdline):
    error = subprocess.run(cmdline, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    print(error.stderr)

def get_changed_registers_info(io, changed_registers):
    registers_string = ' '.join(changed_registers)
    io.sendline(f'info register {registers_string}'.encode())
    output = io.clean().decode()
    lines = output.splitlines()
    if lines and '(gdb)' in lines[-1]:
        lines.pop()
    return '\n'.join(lines).strip()

def test_the_test_cases(io):
    old_registers = parse_registers(DEFAULT_REGISTERS_FILE)
    new_registers = parse_registers(CHANGED_REGISTERS_FILE)
    if ONLY_SHOW_THE_CHANGED_REGISTERS:
        changed_registers = []
        excluded_registers = {'sp', 'fp', 'pc', 'mcycle', 'minstret', 'cycle', 'time', 'instret'}
        for register, new_value in new_registers.items():
            old_value = old_registers.get(register)
            if old_value and register not in excluded_registers and new_value != old_value:
                changed_registers.append(register)
    else:
        changed_registers = CHOSEN_REGISTERS
    if not changed_registers:
        cleaned_output = 'No registers changed!'
    else:
        cleaned_output = get_changed_registers_info(io, changed_registers)
    print(cleaned_output)

def edit_main_s(content):
    with open(code_in_bootloader, 'w', encoding='utf-8') as f:
        f.write(content)

def tidy_output_file(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    start = 0
    while start < len(lines) and not (lines[start].startswith('============') or 'START ITER' in lines[start]):
        start += 1
    kept = lines[start:]
    collapsed = []
    blank_streak = 0
    for ln in kept:
        if ln.strip() == '':
            blank_streak += 1
            if blank_streak <= 1:
                collapsed.append(ln)
        else:
            blank_streak = 0
            collapsed.append(ln)
    while collapsed and collapsed[-1].strip() == '':
        collapsed.pop()
    if collapsed and collapsed[-1].endswith('\n'):
        collapsed[-1] = collapsed[-1].rstrip('\n')
    with open(filename, 'w', encoding='utf-8') as f:
        f.writelines(collapsed)

def run_qemu(function, registers_file, MODE):
    cmdline = ['./run', str(MODE)]
    qemu_io = process(cmdline)
    try:
        qemu_io.recvuntil(b'*** Now run')
        io = process('riscv64-unknown-elf-gdb')
        io.sendline(b'b *test_end')
        io.sendline(b'b *trap_vector')
        io.sendline(b'c')
        output = io.recvuntil(b'Breakpoint 1, ', timeout=TIMEOUT_TIME).decode()
        if not output:
            io.recvuntil(b'Breakpoint 2, ', timeout=TIMEOUT_TIME).decode()
        io.sendline(b'info all-registers')
        output = io.clean().decode()
        registers_cleaned = clean_register_output(output)
        with open(registers_file, 'w', encoding='utf-8') as file:
            file.write(registers_cleaned)
        if function:
            function(io)
        if not output:
            print('ERROR OCCURRED WHILE EXECUTING TEST CASE!')
        io.kill()
    except EOFError:
        handle_qemu_error(cmdline)
    finally:
        qemu_io.kill()

# PATH
DEFAULT_REGISTERS_FILE = 'default_registers.txt'
CHANGED_REGISTERS_FILE = 'changed_registers.txt'
code_in_bootloader = 'bootloader/code.S'
filename_output = 'qemu_result.txt'
input_file = 'qemu_test_cases.txt'

# VARIABLE
DEFAULT_MODE = 3
SKIPPING_N_LINES = 3
TIMEOUT_TIME = 0.1
THE_DEFAULT_STRING = '''li x0, 0
li x1, 0
li x2, 0
li x3, 0
li x4, 0
li x5, 0
li x6, 0
li x7, 0
li x8, 0
li x9, 0
li x10, 0
li x11, 0
li x12, 0
li x13, 0
li x14, 0
li x15, 0
li x16, 0
li x17, 0
li x18, 0
li x19, 0
li x20, 0
li x21, 0
li x22, 0
li x23, 0
li x24, 0
li x25, 0
li x26, 0
li x27, 0
li x28, 0
li x29, 0
li x30, 0
li x31, 0'''
ANSI_ESCAPE = re.compile(r'\x1B\[[0-?]*[ -/]*[@-~]')
ONLY_SHOW_THE_CHANGED_REGISTERS = False
CHOSEN_REGISTERS = ['priv', 'pc', 'x10', 'x11', 'x12', 'mcause']
CODE_TEMPLATE = '''
.globl test_start

test_start:
{test_case}

test_end:
j test_end

.global trap_vector
.align 4

trap_vector:
j trap_vector
'''

# CODE
class Tee:
    def __init__(self, filename):
        self.file = open(filename, 'w')
        self.stdout = sys.stdout
    def write(self, message):
        self.stdout.write(message)
        self.file.write(ANSI_ESCAPE.sub('', message))
    def flush(self):
        self.file.flush()
        self.stdout.flush()
    def close(self):
        self.file.close()

tee = Tee(filename_output)
sys.stdout = tee

with open(input_file, 'r', encoding='utf-8') as file:
    test_cases = file.read()

blocks = [block.strip().split('\n') for block in test_cases.strip().split('\n\n') if block.strip()]
the_list = [[line.strip() for line in block if line.strip()] for block in blocks]
the_list_str = f'the_list = {repr(the_list)}'

exec(the_list_str)

DEFAULT_TEST_CASES = [CODE_TEMPLATE.format(test_case=THE_DEFAULT_STRING)]
CHANGED_TEST_CASES = [CODE_TEMPLATE.format(test_case='\n'.join(THE_DEFAULT_STRING.splitlines() + test_case) if test_case else THE_DEFAULT_STRING) for test_case in the_list]

try:
    edit_main_s(DEFAULT_TEST_CASES[0])
    print('==================\nSTART INITIALIZE 0\n', file=tee.stdout)
    run_qemu(None, DEFAULT_REGISTERS_FILE, DEFAULT_MODE)
    print('\nEND INITIALIZE 0\n================\n', file=tee.stdout)
    for i, (new_content, test_case) in enumerate(zip(CHANGED_TEST_CASES, the_list)):
        edit_main_s(new_content.strip())
        print(f'============\nSTART ITER {i + 1}\n')
        print('\n'.join(test_case) + '\n')
        MODE = DEFAULT_MODE
        run_qemu(test_the_test_cases, CHANGED_REGISTERS_FILE, MODE)
        if i < len(CHANGED_TEST_CASES) - 1:
            print(f'\nEND ITER {i + 1}\n==========\n')
        else:
            print(f'\nEND ITER {i + 1}\n==========')
finally:
    sys.stdout = tee.stdout
    tee.close()
    tidy_output_file(filename_output)

# DONE
print('\n====\nDONE\n====')

# 1 / 1
# result.txt
```
